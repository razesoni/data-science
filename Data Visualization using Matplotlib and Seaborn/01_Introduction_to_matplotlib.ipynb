```python
import matplotlib.pyplot as plt
```

```python
plt.plot([1, 2, 3], [4, 5, 6])
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.plot(years, runs)
plt.xlabel("Years")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar's Yearly Runs")
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
dhoni = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

plt.plot(years, kohli, label="Virat Kohli")
plt.plot(years, dhoni, label="MS Dhoni")

plt.xlabel("Years")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
```

```python
plt.style.use('dark_background')
```

```python
plt.plot(years, kohli, 'ro--', label="Virat Kohli")
plt.plot(years, dhoni, 'g^:', label="MS Dhoni")

plt.xlabel("Years")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.grid(True)
plt.legend()
```
