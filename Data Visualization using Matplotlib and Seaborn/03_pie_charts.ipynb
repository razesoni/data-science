```python
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
```

```python
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [18000, 8000, 12000, 9500]
colors = ['#ff3400','#ff8700','#ff7777','#ff6600' ]

plt.title('Career Runs of Indian Batsmen')
plt.pie(runs, labels=labels, colors=colors)
plt.show()
```

```python
plt.pie(
    runs,
    labels=labels,
    colors=colors,
    wedgeprops={'edgecolor' : '#123456', 'linewidth' : 2, 'linestyle' : '--'}
)
plt.show()
```

```python
explode = [0.1, 0, 0, 0]

plt.title('Exploded pie example')
plt.pie(runs, labels=labels, explode=explode, colors=colors, shadow=True, startangle=120, autopct='%1.1f%%')
plt.show()
```
