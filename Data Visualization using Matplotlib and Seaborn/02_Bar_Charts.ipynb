```python
import matplotlib.pyplot as plt
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]

plt.bar(years, runs)
plt.xlabel("Years")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar's Yearly Runs")
```

```python
import numpy as np

Sachin = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
Kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
Dhoni = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]

x = np.arange(len(years))
width = 0.25

plt.bar(x - width, Sachin, width=width, label="Sachin")
plt.bar(x, Kohli, width=width, label="Kohli")
plt.bar(x + width, Dhoni, width=width, label="Dhoni")

plt.xlabel("Years")
plt.ylabel("Runs")
plt.title("Runs comparison")
plt.xticks(x, years)
plt.legend()
```

```python
players = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
run_5yrs = [500+700+1100+1500+1800, 1200+1700+1300+900+1500, 800+1200+500+800+1100, 1300+1500+1800+1900+2100]
plt.barh(players, run_5yrs, color='skyblue')
plt.xlabel('Total run in first five years')
plt.title("First five year performance of indian batsman")
plt.tight_layout()
```

```python
players = ["Sachin", "Sehwag", "Kohli", "Yuvraj"]
runs = [2000, 3500, 4000, 2500]
plt.bar(players, runs, color='skyblue')

for i in range(len(players)):
    plt.text(i, runs[i] + 30, str(runs[i]), ha='center' )

plt.xlabel('Total run in first five years')
plt.title("First five year performance of indian batsman")
```
