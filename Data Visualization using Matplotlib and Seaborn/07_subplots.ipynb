```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y1 = [i * 2 for i in x]
y2 = [i ** 2 for i in x]
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.title("Double of x")
plt.xlabel("Number")
plt.ylabel("Double of Number")

plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title("Square of x")
plt.xlabel("Number")
plt.ylabel("Square of Number")
plt.tight_layout()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))
axs[0].plot(x, y1)
axs[0].set_title("x * 2")

axs[1].plot(x, y2)
axs[1].set_title("x ** 2")

fig.suptitle("My Comparison plot")
```
