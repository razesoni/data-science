```python
import pandas as pd
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_excel("iris-train.xlsx")
```

```python
data.head()
```

```python
x = data.iloc[:, :-1]
```

```python
x
```

```python
y = data.iloc[:, -1]
```

```python
y
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x, y)
```

```python
model.predict([[5.4, 3, 4.5, 1.5]])
```

```python
datatest = pd.read_excel("iris-test.xlsx")
```

```python
x_test = datatest.iloc[:, :-1]
```

```python
model.predict(x_test.values)
```
