```python
from sklearn.tree import DecisionTreeClassifier
```

```python
features = [[150, 0],
           [170, 0],
           [130, 1],
           [120, 1]]
```

```python
labels = ["apple", "apple", "orange", "orange"]
```

```python
clf = DecisionTreeClassifier()
```

```python
clf = clf.fit(features, labels)
```

```python
clf.predict([[150, 1]])
```
